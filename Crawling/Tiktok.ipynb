```python
#TOTAL
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import threading

# Fungsi untuk memeriksa apakah sudah mencapai bagian bawah halaman
def scroll_down():
    global stop_scroll
    while not stop_scroll:
        driver.execute_script("window.scrollBy(0, 100);")
        time.sleep(0.3)

# Fungsi untuk memantau tombol Enter
def check_enter():
    global stop_scroll
    input("Press Enter to stop scrolling and move to the next page...")
    stop_scroll = True
    print("Moving to the next page...")

# Dictionary untuk menyimpan tautan unik dan data yang telah diambil
unique_links = set()
processed_links = set()
processed_data = set()
post_counter = 226

# Looping untuk melakukan scroll ke dasar halaman dan mengumpulkan tautan yang diinginkan
while True:
    before_scroll_height = driver.execute_script("return document.body.scrollHeight")
    
    all_links = driver.find_elements(By.TAG_NAME, 'a')
    
    for link in all_links:
        href_attribute = link.get_attribute('href')
        if href_attribute and "https://www.tiktok.com/@" and "video" in href_attribute:
            unique_links.add(href_attribute)
    
    driver.find_element(By.TAG_NAME, 'body').send_keys(Keys.END)
    time.sleep(5)

    after_scroll_height = driver.execute_script("return document.body.scrollHeight")

    if after_scroll_height == before_scroll_height:
        break


# Looping untuk setiap link yang telah ditemukan
for link in unique_links:
    if link not in processed_links:
        driver.get(link)
        time.sleep(10)

        stop_scroll = False

        # Membuat dan menjalankan thread untuk memantau tombol Enter
        enter_thread = threading.Thread(target=check_enter)
        enter_thread.start()

        # Memulai proses scroll
        scroll_thread = threading.Thread(target=scroll_down)
        scroll_thread.start()

        driver.execute_script("window.scrollTo(0, 0);")
        time.sleep(5)

        # Dictionary untuk menyimpan data caption dan komentar
        data = {
            "Nama_Caption": [],
            "Isi_Caption": [],
            "Nama_Comment": [],
            "Isi_Comment": []
        }

        while not stop_scroll:
            captions = driver.find_elements(By.CLASS_NAME, "css-15ip68j-DivDescriptionContentWrapper-StyledDetailContentWrapper")
            comments_element = driver.find_elements(By.CLASS_NAME, "css-1i7ohvi-DivCommentItemContainer.eo72wou0")
            printed_comments = set()

            for caption in captions:
                namaCaption = caption.find_elements(By.CLASS_NAME, "css-1c7urt-SpanUniqueId.evv7pft1")
                isiCaption = caption.find_elements(By.CLASS_NAME, "css-1tgwhqp-H1PhotoTitle.eqrezik11")
                caption_number = 1  

                for nama in namaCaption:
                    nama_caption_text = f"{nama.text} {caption_number}"  
                    caption_data = (nama_caption_text, "")  # Membuat tuple untuk caption
                    if caption_data not in processed_data:
                        data["Nama_Caption"].append(nama_caption_text)
                        processed_data.add(caption_data)
                    caption_number += 1  

                for isi in isiCaption:
                    isi_caption_text = isi.text
                    data["Isi_Caption"].append(isi_caption_text)

            for comment in comments_element:
                namaComment = comment.find_elements(By.CLASS_NAME, "css-1665s4c-SpanUserNameText")
                isiComment = comment.find_elements(By.CLASS_NAME, "css-xm2h10-PCommentText.e1g2efjf6")

                for nama in namaComment:
                    nama_comment_text = nama.text
                    comment_data = (nama_comment_text, "")  # Membuat tuple untuk komentar
                    if comment_data not in processed_data:
                        data["Nama_Comment"].append(nama_comment_text)

                for isi in isiComment:
                    isi_comment_text = isi.text
                    comment_data = (nama_comment_text, isi_comment_text)  # Membuat tuple untuk komentar
                    if comment_data not in processed_data:
                        data["Isi_Comment"].append(isi_comment_text)
                        processed_data.add(comment_data)


        file_name = f"post_{post_counter}.txt"
        post_counter += 1
        
        with open(file_name, "w", encoding="utf-8") as text_file:
            text_file.write("Captions:\n")
            for nama_caption_text, isi_caption_text in zip(data["Nama_Caption"], data["Isi_Caption"]):
                text_file.write(f"Nama Caption: {nama_caption_text}\n")
                text_file.write(f"Isi Caption: {isi_caption_text}\n")
                text_file.write("----------\n")
            
            text_file.write("Comments:\n")
            for nama_comment_text, isi_comment_text in zip(data["Nama_Comment"], data["Isi_Comment"]):
                text_file.write(f"Nama Comment: {nama_comment_text}\n")
                text_file.write(f"Isi Comment: {isi_comment_text}\n")
                text_file.write("----------\n")
        
        print(f"Post {post_counter - 1} completed.")
        enter_thread.join()
        scroll_thread.join()

        processed_links.add(link)

print("Done.")
```
